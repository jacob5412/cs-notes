## Hue and Scatter Plots

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue="smoker",
                hue_order=["Yes",
                           "No"])
plt.show()
```

```python
hue_colors = {"Yes": "#808080",
              "No": "#00FF00"}
sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue="smoker",
                palette=hue_colors)
plt.show()
```

```python
hue_colors = {"Male": "#808080",
              "Female": "#00FF00"}
sns.countplot(x = "smoker", data = tips, hue = "sex", palette = hue_colors)
plt.show()
```

## Subplots using `sns.relplot()`

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            col="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            col="smoker",
            row="time")
plt.show()
```

```python
sns.relplot(x="total_bill",
            y="tip",
            data=tips,
            kind="scatter",
            col="day",
            col_wrap=2,
            col_order=["Thur",
                       "Fri",
                       "Sat",
                       "Sun"])
plt.show()
```

## Usage

1. For interactive work, its recommended to use a Jupyter/IPython interface in `matplotlib` mode, or else youll have to call `matplotlib.pyplot.show()` when you want to see the plot:
    - To set this up, before any plotting or import of matplotlib is performed you must execute the `%matplotlib` magic command. This performs the necessary behind-the-scenes setup for IPython to work correctly hand in hand with matplotlib; it does not, however, actually execute any Python import commands, that is, no names are added to the namespace.
    - A particularly interesting backend, provided by IPython, is the `inline` backend. This is available only for the Jupyter Notebook and the Jupyter QtConsole. It can be invoked as follows:
    - the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it

```python
%matplotlib inline
```

2. We apply the default default seaborn theme, scaling, and color palette.
    - uses the [matplotlib rcParam system](https://ipython.readthedocs.io/en/stable/interactive/plotting.html) and will affect how all matplotlib plots look, even if you dont make them with seaborn

```python
sns.set()
```
